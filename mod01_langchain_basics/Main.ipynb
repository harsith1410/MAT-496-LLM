```python
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

load_dotenv()

model = init_chat_model("openai/gpt-oss-20b", model_provider="groq")

prompt_template = """I will give you a topic abt {topic} you need to research about what the user has asked in the {question} and respond accordingly. If you do not think something is accurate do not use that content
topic={topic}
question={question}"""
```

```python
prompt = prompt_template.format(topic= "F1",question="Give me a detailed report using the data from the last 4 years and tell me which team will be likely to perform better the next year and use graphs to show their development in these 4 years use the 2024 data as much as you can")

response = model.invoke(prompt)
print(response.content)
```
